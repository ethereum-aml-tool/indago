```python
from io import StringIO


def read_multiple_files(bucket_name: str, prefix: str, num_files: int) -> pd.DataFrame:
    """
    Reads multiple files from a bucket.
    """
    cols_to_read = ['hash', 'nonce', 'transaction_index', 'from_address',
                    'to_address', 'value', 'gas', 'gas_price', 'block_timestamp', 'block_number']
    bucket = storage_client.get_bucket(bucket_name)
    blobs = list(storage_client.list_blobs(
        bucket_or_name=bucket, prefix=prefix))
    return pd.concat([pd.read_csv(StringIO(blob.download_as_text()), dtype={}, usecols=cols_to_read) for blob in blobs[:num_files]])


files_to_read = 50
#df = read_multiple_files('eth-aml-data', 'transactions', files_to_read)
#df.to_parquet('../data/transactions.csv', index=False, header=True)
df = pd.read_parquet('../data/transactions.parquet')
print(df['block_timestamp'].max())
print(df['block_timestamp'].min())
df
```

```python
kaggle_df = pd.read_csv('../data/kaggle.csv', usecols=['Address', 'Name', 'Label'])
dodgy_kaggle_df = kaggle_df[kaggle_df.Label == 'Dodgy']
matching_transactions = pd.DataFrame(df[df.from_address.isin(dodgy_kaggle_df.Address)], columns=['hash', 'from_address', 'to_address', 'value', 'block_timestamp'])
dodgy_kaggle_df.rename(columns={'Address': 'from_address', 'Name': 'name'}, inplace=True)
names_df = pd.DataFrame(dodgy_kaggle_df, columns=['from_address', 'name'])
matching_transactions = matching_transactions.merge(names_df, on='from_address', how='left')
matching_transactions['value'] = matching_transactions.value.apply(lambda x: int(x) / 10**18)
print(matching_transactions['block_timestamp'].max())
print(matching_transactions['block_timestamp'].min())
matching_transactions.sort_values(by='value', ascending=False, inplace=True)
matching_transactions = matching_transactions[~matching_transactions.to_address.isin(dodgy_kaggle_df.from_address)]
```

```python
import json
df = pd.read_csv('../data/blacklist_results/fifo.csv')
df['queue'] = df['queue'].map(lambda x: '{' + x.replace('tainted:', '"tainted":').replace('untainted', '"untainted"').replace('un"ta', '"unta').replace(':', ' :') + '}')
df['queue'] = df['queue'].map(lambda x: json.loads(x))
df2 = pd.json_normalize(df['queue'])
df = pd.concat([df.drop(['queue'], axis=1), df2], axis=1)
df.dropna(inplace=True)
df.to_csv('../data/blacklist_results/fifo.csv', index=False)
df
```
